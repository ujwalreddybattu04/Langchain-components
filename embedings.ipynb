```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings = OllamaEmbeddings(model="nomic-embed-text")

r1 = embeddings.embed_documents([
    "alpha is the first letter of greek",
    "beta is the second letter of greek",
])
print(r1)
```

```python
embeddings.embed_query("this is the third letter of greek")
```
